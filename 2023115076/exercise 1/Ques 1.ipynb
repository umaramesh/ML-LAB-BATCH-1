```python
import numpy as np
from numpy.linalg import det,eig
A=np.array([[4,2],[3,1]])
print("matrix:\n",A)
print("determinent of arr: ",det(A))
eigenvalues,eigenvectors=eig(A)
print("\nEigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
```

```python
import numpy as np
my_list=[1,3,5,2,4,6,8,7]
print("list: ",my_list)
arr_1d=np.array(my_list)
print("1d array: ",arr_1d)
arr_2d=np.array(my_list).reshape(4,2)
print("2d array: ",arr_2d)
arr_3d=np.array(my_list).reshape(2,2,2)
print("3d array: ",arr_3d)
```

```python
import numpy as np
print("NumPy Array and Matrix Generation:\n")
arr=np.arange(1,11)
print("arange(1,11):",arr)
ones_matrix=np.ones((3,3))
print("\n3*3 ones matrix:\n",ones_matrix)
zeros_matrix=np.zeros((3,3))
print("\n3*3 zeros matrix:\n",zeros_matrix)
identity_matrix=np.eye(3)
print("\n3*3 identity matrix:\n",identity_matrix)
random_matrix = np.random.rand(2, 3)
floor_matrix = np.floor(random_matrix * 10)
print("\n2x3 Random Matrix [0,1]:\n", floor_matrix)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print("\nSeries created from a list:")
print(series)

print("\nAccessing index and values of the Series:")
print("Index:", series.index.tolist())
print("Values:", series.values)

print("\nCompare numpy Array vs panda Series:")
np_array=np.array({10,20,30,40,50})
print("numpy array: ",type(np_array))
print("panadas series type: ",type(series))

print("\nsSeries with custom indices:")
named_series=pd.Series([90,80,70],index=["math","science","english"])
print(named_series)

print("\naccess Single value by index:\n",named_series)

print("\nDataFrame:")
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Marks': [85, 92, 78]
}
df = pd.DataFrame(data)
print(df)

plt.figure(figsize=(5,3))
plt.bar(df['Name'], df['Marks'], color='skyblue')
plt.title("Student Marks Bar Chart")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```
