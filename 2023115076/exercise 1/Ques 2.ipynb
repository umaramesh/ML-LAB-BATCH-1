Implementation of Python Libraries for ML application such as Pandas and Matplotlib.
a) Create a Series using pandas and display


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Create Series using Pandas:")
series = pd.Series([5, 10, 15, 20, 25])
print(series)
```

b) Access the index and the values of our Series


```python
print("\nAccess Index and Values:")
print("Index:", series.index.tolist())
print("Values:", series.values)
```

c) Compare an array using Numpy with a series using pandas


```python
print("\nCompare Numpy Array with Pandas Series:")
np_array = np.array([5, 10, 15, 20, 25])
print("NumPy Array Type:", type(np_array))
print("Pandas Series Type:", type(series))
```

d) Define Series objects with individual indices


```python
print("\nSeries with Custom Index:")
custom_series = pd.Series([80, 85, 90], index=["Math", "Science", "English"])
print(custom_series)
```

e) Access single value of a series


```python
print("\nAccess Single Value from Series (Math):")
print("Marks in Math:", custom_series["Math"])
```

f) Load datasets in a Data frame variable using


```python
print("\nDataFrame Example (Student Dataset):")
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 92, 78, 88]
}
df = pd.DataFrame(data)
print(df)
```
