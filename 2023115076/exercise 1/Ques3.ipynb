 Creation and Loading different types of datasets in Python using the required libraries.

i. Creation using pandas

ii. Loading CSV dataset files using Pandas

iii. Loading datasets using sklearn


```python
import pandas as pd
from sklearn import datasets

print("i. Creating Dataset using Pandas:")
student_data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [22, 23, 21],
    'Score': [85, 90, 78]
}
df_students = pd.DataFrame(student_data)
print(df_students)

print("\nii. Loading CSV Dataset using Pandas:")
df_iris = pd.read_csv("iris_downloaded.csv")
print(df_iris.head())

print("\niii. Loading Dataset using sklearn (Iris Example):")

iris = datasets.load_iris()
print("Dataset Keys:", iris.keys())

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
print(iris_df.head())
```

```python
import pandas as pd
import statistics as stats

data = {'Marks': [85, 90, 78, 92, 85, 88, 78, 90, 85, 92]}
df = pd.DataFrame(data)

print("Dataset (Marks):")
print(df)

marks = df['Marks'].tolist()

mean_value = stats.mean(marks)
median_value = stats.median(marks)
mode_value = stats.mode(marks)
variance_value = stats.variance(marks)
std_deviation = stats.stdev(marks)

print("\nStatistical Calculations:")
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_deviation)
```
