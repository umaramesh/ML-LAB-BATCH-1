```python
import pandas as pd
import statistics as stats
data = {'Marks': [85, 90, 78, 92, 85, 88, 78, 90, 85, 92]}
df = pd.DataFrame(data)

print("Dataset (Marks):")
print(df)

marks = df['Marks'].tolist()

mean_value = stats.mean(marks)
median_value = stats.median(marks)
mode_value = stats.mode(marks)
variance_value = stats.variance(marks)
std_deviation = stats.stdev(marks)

print("\nStatistical Calculations:")
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_deviation)
```

Demonstrate various data pre-processing techniques for a given dataset.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

data = {'Marks': [85, 90, 78, 92, 85, 88, 78, 90, 85, 92]}
df = pd.DataFrame(data)
print("Original Dataset:\n", df)
df.loc[2, 'Marks'] = np.nan
print("\nDataset with Missing Value:\n", df)

df['Marks'].fillna(df['Marks'].mean(), inplace=True)
print("\nAfter Filling Missing Values with Mean:\n", df)

df = df.drop_duplicates()
print("\nAfter Removing Duplicates:\n", df)
scaler = MinMaxScaler()
df['Marks_Normalized'] = scaler.fit_transform(df[['Marks']])

std_scaler = StandardScaler()
df['Marks_Standardized'] = std_scaler.fit_transform(df[['Marks']])

print("\nAfter Feature Scaling (Normalized and Standardized):\n", df)

print("\nFinal Preprocessed Dataset:\n", df)
```

 (C) Demonstrate various data pre-processing techniques for a given dataset. Write a python program to compute
i. Reshaping the data,
ii. Filtering the data,
iii. Merging the data
iv. Handling the missing values in datasets
v. Feature Normalization: Min-max normalization

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

data1 = {
    'ID': [1, 2, 3, 4, 5],
    'Marks': [85, 90, np.nan, 92, 88],
    'Age': [21, 22, 23, np.nan, 21]
}
df1 = pd.DataFrame(data1)
print("Dataset 1:\n", df1)

print("\nReshaping Data using melt():")
reshaped_df = df1.melt(id_vars=['ID'], var_name='Variable', value_name='Value')
print(reshaped_df)

print("\nFiltering Data (Marks > 85):")
filtered_df = df1[df1['Marks'] > 85]
print(filtered_df)

data2 = {
    'ID': [3, 4, 5, 6],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata']
}
df2 = pd.DataFrame(data2)
print("\nDataset 2:\n", df2)

merged_df = pd.merge(df1, df2, on='ID', how='left')
print("\nAfter Merging Dataset 1 and 2:\n", merged_df)

print("\nHandling Missing Values (fillna):")
merged_df['Marks'].fillna(merged_df['Marks'].mean(), inplace=True)
merged_df['Age'].fillna(merged_df['Age'].median(), inplace=True)
print(merged_df)

print("\nMin-Max Normalization on Marks and Age:")
scaler = MinMaxScaler()
merged_df[['Marks_Normalized', 'Age_Normalized']] = scaler.fit_transform(merged_df[['Marks', 'Age']])
print(merged_df[['ID', 'Marks_Normalized', 'Age_Normalized']])

print("\nFinal Preprocessed Dataset:\n", merged_df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, Binarizer
df = pd.read_csv("sample.csv")

label_encoder = LabelEncoder()
df['gender_encoded'] = label_encoder.fit_transform(df['gender'])
df['selected_encoded'] = label_encoder.fit_transform(df['selected'])
print("\nAfter Label Encoding:\n", df[['gender', 'gender_encoded' , 'selected', 'selected_encoded']])
scaler = MinMaxScaler()
df['Marks_scaled'] = scaler.fit_transform(df[['Marks']])
print("\nAfter Scaling Marks (MinMaxScaler):\n", df[['Marks', 'Marks_scaled']])
binarizer = Binarizer(threshold=50)
df['Marks_binarized'] = binarizer.fit_transform(df[['Marks']])
print("\nAfter Binarizing Marks (threshold = 50):\n", df[['Marks', 'Marks_binarized']])

print("\nFinal Preprocessed Dataset:\n", df)
```
