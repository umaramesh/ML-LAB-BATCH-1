```python
import math

print("The Ceil of 3.2 is :",math.ceil(3.2))
print("The Floor of 3.2 is :",math.floor(3.2))
print("The Square Root of 9 is :",math.sqrt(9))
print("The Interger Square Root of 3.2 is :",math.isqrt(9))
print("The GCD of 3 and 9 is :",math.gcd(3,9))
print("6C4 is :",math.comb(6,4))
print("6P4 is :",math.perm(6,4))
print("Log of 10 base 10 is :",math.log(10,10))
```

```python
import numpy as np
import statistics as st

arr=np.array([44,23,67,89,12,45])
print("Mean :",st.mean(arr))
print("Mean :",np.mean(arr))
print("Median :",st.median(arr))
print("Median :",np.median(arr))
print("Mode :",st.mode(arr))
print("Variance :",np.var(arr))
# print("Standard deviation :",st.stdev(arr))
print("Standard deviation :",np.std(arr))
```
