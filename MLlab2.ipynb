```python
import numpy as np
from scipy import stats

data = [12, 18, 25, 25, 30, 30, 30, 35, 40, 50]
arr = np.array(data)

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Mode:", stats.mode(arr, keepdims=True).mode[0])
print("Variance:", np.var(arr))
print("Standard Deviation:", np.std(arr))
```

```python
df = pd.DataFrame({'Name': ['A', 'B', 'C'], 'Marks': [85, 60, 95]})
filtered = df[df['Marks'] > 70]
print("Filtered Data:\n", filtered)
```

```python
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['A', 'B']})
df2 = pd.DataFrame({'ID': [1, 2], 'City': ['Chennai', 'Madurai']})
merged = pd.merge(df1, df2, on='ID')
print("Merged Data:\n", merged)
```

```python
df = pd.DataFrame({
    'Name': ['A', 'B', 'C'],
    'Age': [25, None, 30],
    'Marks': [85, 90, None]
})
print("Original Data:\n", df)
df_filled = df.fillna(df.mean(numeric_only=True))
print("After Handling Missing Values:\n", df_filled)
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame({'Age': [20, 25, 30, 35], 'Marks': [50, 60, 80, 90]})
scaler = MinMaxScaler()
df[['Age', 'Marks']] = scaler.fit_transform(df[['Age', 'Marks']])
print("After Min-Max Normalization:\n", df)
```
