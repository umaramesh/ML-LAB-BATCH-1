```python
import pandas as pd

# Step 1: Create data
data = {
    "Size": [1000, 1500, 2000, 2500, 3000],
    "Price": [50, 70, 90, 110, 130]
}

# Step 2: Convert to DataFrame
df = pd.DataFrame(data)

# Step 3: Save as CSV (in same folder as this notebook)
df.to_csv("house_data.csv", index=False)
print("CSV file 'house_data.csv' created successfully!")

# Step 4: Display the data
df
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("house_data.csv")

# Show the first few rows
print("Loaded CSV data:")
df.head()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Separate features (X) and target (y)
X = df[["Size"]]  # Feature: house size
y = df["Price"]   # Target: house price

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Show slope and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)

# Predict price for 2200 sq ft
predicted_price = model.predict([[2200]])
print("Predicted price for 2200 sq ft:", predicted_price[0])
```

```python
# Plot actual data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot best fit line
plt.plot(X, model.predict(X), color='red', label='Best Fit Line')

plt.xlabel("House Size (sq ft)")
plt.ylabel("Price (â‚¹ lakhs)")
plt.title("Linear Regression Example")
plt.legend()
plt.show()
```
