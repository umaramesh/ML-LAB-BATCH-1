```python
import math

print("Floor of 3.2 is ",math.floor(3.2))
print("Ceil of 3.2 is ",math.ceil(3.2))
print("Square root of 4 is ",math.sqrt(4))
print("Integer Square root of 4 is ",math.isqrt(4))
print("GCD of 3 and 2 is ",math.gcd(3,2))
```

```python
import numpy as np

arr=np.array([[1,2,3],[4,5,6],[8,7,9]])

print("Dimensions :",arr.ndim)
print("Shape :",arr.shape)
print("Size :",arr.size)
print("Sum of elements :",arr.sum())
print("Sum of elements :",sum(arr)) #sum of elements of the column
print("Mean of the array :",arr.mean())
print("Original array :\n",arr)
print("Sorted array :\n",np.sort(arr))
print("Sine of the array :\n",np.sin(arr))
```

```python
matrix=np.array([[1,2],[7,8]])

det_val=np.linalg.det(matrix)

print("The Determinant of the matrix is :",det_val)

eig_val,eig_vector=np.linalg.eig(matrix)

print("Eigen Value :",eig_val)
print("Eigen Vector :\n",eig_vector)
```

```python
list=[i for i in range(1,13)]

array1D=np.array(list)
array2D=array1D.reshape(4,3)
array3D=array1D.reshape(2,2,3)

print("List :",list)
print("1D array :",array1D)
print("2D array :\n",array2D)
print("3D array :\n",array3D)
```

```python
zero_mat=np.zeros((3,3))
ones_mat=np.ones((3,3))
random_mat=np.random.rand(3,3)
identity_mat=np.eye(3)

print("Zero matrix :\n",zero_mat)
print("Ones matrix :\n",ones_mat)
print("Random matrix :\n",random_mat)
print("Identity matrix :\n",identity_mat)
```

```python
from scipy.linalg import det

arr=np.array([[1,0,0],[1,1,0],[0,0,0]])
det_val=det(arr)
print("The determinant of the array is :",det_val)
```

```python
from scipy.linalg import eig
matrix=np.array([[1,2],[7,8]])
eig_val,eig_vector=eig(matrix)

print("Eigen Value :",eig_val)
print("Eigen Vector :\n",eig_vector)
```
