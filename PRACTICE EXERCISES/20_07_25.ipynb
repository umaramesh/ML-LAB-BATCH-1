```python
from scipy import linalg

matrix=np.array(([4,5,6],[1,2,3],[7,8,9]))
print("Matrix :\n",matrix)

det=linalg.det(matrix)
print("Determinant of the matrix :",det)
```

```python
eig_val,eig_vec=linalg.eig(matrix)
print(f"Eigen Value : {eig_val} \n Eigen Vector : \n{eig_vec}")
```

```python
from sklearn.datasets import load_iris

iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
print(df.head)
```

```python
from scipy import stats

print("Mean : \n",df.mean())
print("\nMedian : \n",df.median())
print("\nMode : \n",df.mode().iloc[0])
print("\nVariance : \n",df.var())
print("\nStandard Deviation : \n",df.std())
```

```python
df_long=pd.melt(df,id_vars=['target'],var_name='Features',value_name='Value')
print(df_long.head())
```

```python
filtered_df=df[df['sepal length (cm)'] > 5.0]
print(filtered_df.head())
```
